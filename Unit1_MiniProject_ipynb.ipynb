# Unit 1 Mini Project – Text Summarization

## Problem Statement
Automatic text summarization using a Transformer-based encoder-decoder model.

## Model Used
facebook/bart-large-cnn

## Description
This mini project demonstrates the use of an encoder-decoder architecture for abstractive text summarization using a pre-trained BART model.


```python
from transformers import pipeline

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Input text
text = """
Generative Artificial Intelligence is transforming multiple industries by enabling machines
to generate text, images, and other content. However, it also introduces challenges such as
hallucinations, bias, and ethical concerns.
"""

# Generate summary
summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

summary
```

## Observation

The BART model successfully generated a concise and meaningful summary of the input text.
This demonstrates the effectiveness of encoder-decoder Transformer architectures for
sequence-to-sequence tasks like summarization.
